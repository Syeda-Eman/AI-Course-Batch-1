## **Operators in Python**

Python operators are symbols that perform an operation on one or more operand. An operand is a variable or a value on which we perform the operations.

Operators can manipulate individual items and return a result. Operators manipulate individual items by conducting mathematical operations and/or returning a Boolean value (True or False)

Mathematical operators are basic arithmetic symbols used to perform addition (+), subtraction (-), multiplication (*), division(/), modulus(%) and many more.

```python
3+2
```

```python
3-2
```

```python
3*2
```

```python
3/2
```

```python
3**2
```

```python
11%2
```

```python
11//4
```

```python
#you can also perform arithmetics like this:
print(3+2)

a= 5
b= 2
print(a+b)

length = 2.5
height = 3.5
area = length*height
print(area)
```

## **Comparison Operators**
Comparison operators are used to compare values, they either return True or False.

```python
a==a
```

```python
a!=b
```

```python
2>1
```

```python
1<2
```

```python
a = 300
b = 250

result = a == b
print("Is a equal to b?", result)
```

```python
result = a != b
print("Is a not equal to b?", result)
```

```python
result = a>b
```

## **Membership Operators**
Membership operators are used to test whether a value or a variable is found in a sequence (string, tuple, list, set, or dictionary)
These operators become essential when we discuss loops.

```python
'cat' in ['dog','cat','mouse']
```

```python
'geese' not in ['dog','cat','mouse']
```
