## **Python While Loops**
Loops are used in programming to repeat a specefic block of code.
The WHILE loop is used when we want to repeat a block of code as long as the statement given is True.

When the condition becomes False, execution stops immediately and if the statement can never be false again then your loop will become infinite.

We generally use this loop when we don't know the number of times to iterate beforehand.

Python interprets any non-zero as True. None and 0 are interpreted as False.



# **When to use While loop**


*   **Automate and repeat tasks:** While loops execute blocks of code over and over again untill the condition is met. It allows us to automate and repeat tasks in an effecient manner.
*   **Indefinite Iteration:** The while loop will run as often as necessary to complete a particular task. When the user doesn't know the number of iterations before execution, while loop is used instead of a loop.
* **Reduce complexity:** While loopa are easy to write. While using WHILE loops, we don’t need to write the statements again and again. Instead, we can write statements we want to execute repetively inside the body of the loop thus, reducing the complexity of the code.
* **Infinite loop:** If the code inside the **`while`** loop doesn’t modify the variables being tested in the loop condition, the loop will run forever.

6. Quadratic equation is calculated as follows: ax²+bx+c=0. Write a function called solve_quadratic_eqn which calculates the solution set of a quadratic equation.

```python
def solve_quadratic_eqn(a, b, c):
  discriminant = b**2 - 4*a*c
  if discriminant < 0:
    return ("No real roots.")
  elif discriminant == 0:
    return(- b/(2*a),)
  else:
    return (-b + (discriminant**0.5))/(2*a), (-b - (discriminant**.5))/2/a


roots = solve_quadratic_eqn(1, -2, 1)
if roots:
  print(f"The roots of the equation are: {roots}")
else:
  print("The equation has no real roots.")
```

1. Declare a function named evens_and_odds. Which should take a positive integer as parameter and count the number of evens and odds till that number.

```python
def evens_and_odds(num):
    even_nums = 0
    odd_nums = 0
    for i in range(1, num+1):
      if i % 2 == 0:
          even_nums += 1
      else:
          odd_nums += 1
    return even_nums, odd_nums

evens_and_odds(100)
```

4. Write different functions which take lists. They should calculate_mean, calculate_median, calculate_mode, calculate_range, calculate_variance, calculate_std(standard deviation).

```python
import statistics

def calculate_mean(value):
  return statistics.mean(value)

def calculate_median(value):
  return statistics.median(value)

def calculate_mode(value):
  return statistics.mode(value)

def calculate_range(value):
  return max(value) - min(value)

def calculate_variance(value):
  return statistics.variance(value)

def calculate_std(value):
  return statistics.stdev(value)


value = [1,2,3,4,5]

print(calculate_mean(value))
print(calculate_median(value))
print(calculate_mode(value))
print(calculate_range(value))
print(calculate_variance(value))
print(calculate_std(value))
```

```python
def is_prime(num):
    return num > 1 and all(num % prime_num for prime_num in range(2, int(num**0.5)+1))

num = 12
prime_num = is_prime(num)
print(prime_num)
```
