## **3-Tuple**
Tuples are used to store multiple items in a single variable.
A tuple is a collection which is ordered, unchangeable and dont allow duplicates.

The first item has index[0], the second item has index[1] etc.

Tuples are written with round brackets ().

```python
thistuple = ("apple","banana","cherry")
print(thistuple)
```

## **Creating Tuple with one item**
To create a tuple with only one item, always remember to add a comma after the item, otherwise it will not be recognised as a tuple.

```python
thistuple = ("apple",)
print(type(thistuple))

thistuple = ("orange")
print(type(thistuple))
```

## **Tuple Items - Data Types**
Tuple items can be of any data type

```python
#String, Integer, and Boolean:
tuple1 = ("apple", "banana", "cherry")
tuple2 = (1,5,7,8,3)
tuple3 = (True, False, False)
tuple4 = ("abc", 35, True)
```

## **4-Sets**
A set is collection which is unordered, unchangeable, unindexed and dont allow duplicates. They are used to store multiple items in a single variable.

They are written with curly brackets {}.

**Note:** The set is unchangeable but still you can add and remove items from it.

```python
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)
```

## **Set Items - Data Types**
Set items can be of any data type:

```python
#String, Integer, and Boolean
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 8, 9}
set3 = {True, False, True}
set4 = {"abc", 34, True, 40, "male"}
```

```python
print(set1)
print(set2)
print(set3)
print(set4)
```

## **5- Dictionaries**

A dictionary is a collection which is used to store data values in "key:value" pairs, and can be referred to by using the key name.

It is unordered, changeable, and do not allow duplicates.

Dictionaries are written with curly brackets, and have keys and values.

```python
thisdict = {"brand": "Ford", "model": "Mustang", "year": 1964}
print(thisdict)
```

```python
thisdict = {
    "brand":"Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict["brand"])
```

```python
thisdict["brand"]="Tesla"
print(thisdict)
```

```python
#this will overwrite the existing value and will write the latest value
thisdict = {
    "brand":"Ford",
    "model": "Mustang",
    "year": 1964,
    "year": 2020
}
print(thisdict)
```

## **Dictionary Items - Data Types**
Every key can have the same value regardless of their datatype.
